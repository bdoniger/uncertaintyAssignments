Fair 3-sided die:
- Outcomes: (1,2,3)
- Set of all events: $\{\varnothing, \{1\}, \{2\}, \{3\}, \{1, 2\}, \{2, 3\}, \{1, 3\}, \{1, 2, 3\} \}$
- Probabilities:
    - $P(\varnothing) = 0$, $P(\{1,2,3\}) = 1$,
    - $P(1) = P(2) = P(3) = \frac{1}{3}$, 
    - $P(\{1,2\}) = P(\{2,3\}) = P(\{1,3\}) = \frac{2}{3}$

Flipping a fair coin twice:
- Outcomes: (HH, HT, TH, TT)
- Set of all events: $\{\varnothing, \{HH\}, \{HT\}, \{TH\}, \{TT\}, \{HH, HT\}, \{HH, TH\}, \{HH, TT\}, \{HT, TH\}, \{HT, TT\}, \{TH, TT\},$
$ \{HH, HT, TH\}, \{HH, HT, TT\}, \{HH, TH, TT\}, \{HT, TH, TT\}, \{HH, HT, TH, TT\} \}$
- Probabilities
    - $P(\varnothing) = 0$, $P(\{HH, HT, TH, TT\}) = 1$, 
    - $P(HH) = P(HT) = P(TH) = P(TT) = \frac{1}{4}$, 
    - $P(\{HH, HT\}) = P(\{HH, TH\}) = P(\{HH, TT\}) = P(\{HT, TH\}) = P(\{HT, TT\}) = P(\{TH, TT\}) = \frac{1}{2}$
    - $P(\{HH, HT, TH\}) = P(\{HH, HT, TT\}) = P(\{HH, TH, TT\}) = P(\{HT, TH, TT\}) = \frac{3}{4}$

6 sided die:

- Probabilities:

```python
#random var that assigns the squre root of the # of pips on the die to each outcome
import random
import numpy as np

def roll_die():
    return random.randint(1,6)

def sim_rolls(n):
    return [roll_die() for _ in range(n)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y = np.sqrt(sim_rolls(5000))
sns.ecdfplot(y).set(title='ECDF, sqrt of die roll (n=5000)')
plt.show()
sns.kdeplot(y).set(title='KDE, sqrt of die roll (n=5000)')
plt.show()
```

- Logistic distribution (similar to normal): 
$$
F(x; \sigma) = \dfrac{1}{1+e^{-x/\sigma}} 
$$
with $\sigma >0$.


- Support: Real numbers
- density: $ \frac{e^{\frac{-x}{\sigma}}}{\sigma (e^{\frac{-x}{\sigma}}+1)^2} $

```python
import scipy as sp
import pandas as pd
```

```python
#parameters
mu = 0
sigma = [0.5, 1, 2, 5]

grid = np.linspace(-10,10,200)

fig, axes = plt.subplots(1,2, figsize=(10,5))

for s in sigma:
    pdf = sp.stats.logistic.pdf(grid, scale=s, loc=mu)
    cdf = sp.stats.logistic.cdf(grid, scale=s, loc=mu)
    sns.lineplot(x=grid,y=pdf, ax=axes[0], label=f'sigma={s}')
    axes[0].set_title(f'PDF, Logistic, sigma={s}')
    sns.lineplot(x=grid,y=cdf, ax=axes[1], label=f'sigma={s}')
    axes[1].set_title(f'CDF, Logistic, sigma={s}')
```

```python
# code based off notes in 01_zoo.ipynb
# sampling
sample = sp.stats.logistic.rvs(scale=1, loc=mu, size=1000)
pdf = sp.stats.logistic.pdf(grid, scale=1, loc=mu)
cdf = sp.stats.logistic.cdf(grid, scale=1, loc=mu)

fig, axes = plt.subplots(1,2, figsize=(14,5))
sns.kdeplot(sample, ax=axes[0], label='kde')
sns.lineplot(x=grid, y=pdf, ax=axes[0], label='pdf', linestyle='--')
axes[0].set_title(f'Sample KDE')

sns.ecdfplot(sample, ax=axes[1], label='ecdf')
sns.lineplot(x=grid, y=cdf, ax=axes[1], label='cdf', linestyle='--')
axes[1].set_title(f'Sample ECDF')

plt.show()
```

- Exponential distribution (similar to log-normal):
$$
F_X(x) = \begin{cases}
0, & x<0 \\
1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.


- Support: $[0, \infty]$
- density: $ \lambda e^{-\lambda x} $

```python
#parameters
mu = 0
mylambda = [0.5, 1, 2, 5]

grid = np.linspace(0,50,200)

fig, axes = plt.subplots(1,2, figsize=(10,5))

for l in mylambda:
    pdf = sp.stats.expon.pdf(grid, scale=l, loc=mu)
    cdf = sp.stats.expon.cdf(grid, scale=l, loc=mu)
    sns.lineplot(x=grid,y=pdf, ax=axes[0], label=f'lambda={l}')
    axes[0].set_title(f'PDF, Exponential')
    sns.lineplot(x=grid,y=cdf, ax=axes[1], label=f'lamda={l}')
    axes[1].set_title(f'CDF, Exponential')
```

```python
# code based off notes in 01_zoo.ipynb
# sampling
sample = sp.stats.expon.rvs(scale=1, loc=mu, size=1000)
pdf = sp.stats.expon.pdf(grid, scale=1, loc=mu)
cdf = sp.stats.expon.cdf(grid, scale=1, loc=mu)

fig, axes = plt.subplots(1,2, figsize=(14,5))
sns.kdeplot(sample, ax=axes[0], label='kde')
sns.lineplot(x=grid, y=pdf, ax=axes[0], label='pdf', linestyle='--')
axes[0].set_title(f'Sample KDE')

sns.ecdfplot(sample, ax=axes[1], label='ecdf')
sns.lineplot(x=grid, y=cdf, ax=axes[1], label='cdf', linestyle='--')
axes[1].set_title(f'Sample ECDF')

plt.show()
```

- Negative Binomial (similar to Poisson): The probability mass function for positive integers is:
$$
f(k;r,p) = \dfrac{(k+r-1)!}{k!(r-1)!}(1-p)^k p^r, \quad \text{ for $k=0,...n$}
$$
You can interpret this as follows: Flip a coin that comes up heads with probability $p$ until you get $r$ heads, and then stop. What is the probability of stopping at each $k=0,1,2,...$?


Support: [0, 1, 2, ..., n]

```python
#parameters
n = 10
prob = [0.2, 0.4, 0.6, 0.8]

x = np.linspace(-1,10,100)

fig, axes = plt.subplots(1,2, figsize=(10,5))

for p in prob:
    pmf = sp.stats.nbinom.pmf(x, n, p, loc=mu)
    cdf = sp.stats.nbinom.cdf(x, n, p, loc=mu)
    sns.lineplot(x=x,y=pmf, ax=axes[0], label=f'p={p}')
    axes[0].set_title(f'PMF, Negative Binomial')
    sns.lineplot(x=x,y=cdf, ax=axes[1], label=f'p={p}')
    axes[1].set_title(f'CDF, Negative Binomial')
```

```python
# code based off notes in 01_zoo.ipynb
# sampling
p = 0.5
n = 10

sample = sp.stats.nbinom.rvs(n, p, loc=mu, size=1000)
pmf = sp.stats.nbinom.pmf(x, n, p, loc=mu)
cdf = sp.stats.nbinom.cdf(x, n, p, loc=mu)

fig, axes = plt.subplots(1,2, figsize=(14,5))
sns.histplot(x=sample,ax=axes[0],label='kde')
sns.lineplot(x=x,y=pmf,ax=axes[0],label='pmf',linestyle='--')
axes[0].set_title('Sample PMF')


sns.ecdfplot(sample, ax=axes[1], label='ecdf')
sns.lineplot(x=x, y=cdf, ax=axes[1], label='cdf', linestyle='--')
axes[1].set_title(f'Sample ECDF')

plt.show()
```

- Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

Support: {1,2, .... k}

```python
cats = ['A', 'B', 'C', 'D']
probs = [0.2, 0.4, 0.1, 0.3]

samples = np.random.choice(cats, size=1000, p=probs)
fig, axes = plt.subplots(1,2, figsize=(14,5))
sns.histplot(samples, ax=axes[0],label='kde').set(title='PMF')

sns.ecdfplot(samples, ax=axes[1],label='ecdf').set(title='ECDF')
```
